```python
def insertionSort(arr):
    sorted_arr=[arr[0]]
    
    for element in arr[1:]:
        l=len(sorted_arr)
        for i in range(l):
            if element<sorted_arr[i]:
                sorted_arr=sorted_arr[:i]+[element]+sorted_arr[i:]
                break
        
        else:
            sorted_arr=sorted_arr+[element]
            
    return sorted_arr
                
```

```python
def insertionSort2(arr):
    sorted_arr=[arr[0]]
   
    for element in arr[1:]:

        for i in range(len(sorted_arr)):
            if element>sorted_arr[-1]:
                sorted_arr=sorted_arr+[element]
                break 
            elif element>sorted_arr[-i-1]:
                sorted_arr=sorted_arr[:-i]+[element]+sorted_arr[-i:]
                
                break
        
        else:
                sorted_arr=[element]+sorted_arr
                          
            
    return sorted_arr
                
```
